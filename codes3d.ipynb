# 2. Heatmap

```r
summ = read.table('codes3d/heatmap.tsv',sep='\t',head=T)
#summ[summ==0] <- NA
print(dim(summ)); head(summ)
```

```r
library(ComplexHeatmap)
library(circlize)
library(dendextend)
```

```r
mat   = as.matrix(summ[,-1])
row.names(mat) = summ[,1]
#mat.s = t(scale(t(mat)))
mat.s = mat
summary(mat.s)
```

```r
mat   = as.matrix(summ[,-1])
row.names(mat) = summ[,1]
mat.s = t(scale(t(mat)))

mat.s_max = max(mat.s,na.rm=T); mat.s_min = min(mat.s,na.rm=T)
#col.rg = c(mat.s_min,0,mat.s_max)
#col.rg = c(-3,0,5)
#cell.cols = colorRamp2(col.rg,c("Cyan","black","Yellow"))
#col.rg = c(-1,1,3,5)
#cell.cols = colorRamp2(col.rg,c("black", "cornflowerblue", "yellow", "red"))
col.rg = c(-1,3,5)
cell.cols = colorRamp2(col.rg,c("black", "cornflowerblue", "yellow"))

hr = hclust(dist(mat.s),method="average")
clu = as.data.frame(cutree(hr,h=8))
clu[clu==3] <- 'C1'
clu[clu==8] <- 'C2'
clu[clu==10]<- 'C3'
clu[clu==17]<- 'C4'
clu[clu==11]<- 'C5'
clu[clu==4] <- 'C6'
clu[clu==5] <- 'C7'
clu[clu==2] <- 'C8'
clu[clu==1] <- 'Minor'
clu[clu==6] <- 'Minor'
clu[clu==7] <- 'Minor'
clu[clu==9] <- 'Minor'
clu[clu==12]<- 'Minor'
clu[clu==13]<- 'Minor'
clu[clu==14]<- 'Minor'
clu[clu==15]<- 'Minor'
clu[clu==16]<- 'Minor'
clu[clu==18]<- 'Minor'
clu[clu==19]<- 'Minor'
h=Heatmap(mat.s,
          column_title='',
          name='GTEx\nMedian\nTPM\n(Z-scaled)',
          #km=10, # k-means cluster
          #cluster_rows=clusters,
          split=clu,
          cluster_column=F,
          rect_gp=gpar(col='black'),
          col=cell.cols)
print(h)
#dev.copy(png,'codes3d/73 Genes from CoDeS3D.png',width=12,height=16,units='in',res=100)
dev.copy(svg,'codes3d/73 Genes from CoDeS3D.svg',width=12,height=16)
while(!is.null(dev.list())) dev.off()
```
